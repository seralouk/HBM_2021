## LOOCV plot results for Reviewer 2, round 2

```python
import pandas as pd
import numpy as np, matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_to_results = "/Users/loukas/Desktop/RS_DATA/Pipeline/Report_results/RS_paper/PLSC/LOOCV/"
```

```python
Lx_full = pd.read_csv(path_to_results + "Lx_full.csv", header=None).values.ravel()
Ly_full = pd.read_csv(path_to_results + "Ly_full.csv", header=None).values.ravel()

# CV results
Lx_CV = pd.read_csv(path_to_results + "Lx_CV.csv", header=None).values.ravel()
Ly_CV = pd.read_csv(path_to_results + "Ly_CV.csv", header=None).values.ravel()
```

```python
Lx_full, Lx_CV
```

```python
plt.figure()
ax=sns.regplot(x=Lx_full, y=Lx_CV, color="b", x_ci=0, fit_reg = 1, scatter_kws={"color": "black"}, line_kws={"color": "royalblue"});
plt.xlabel("Brain scores (full)")
plt.ylabel("LOOCV Brain scores")
plt.xlim([-12.5,5.5])
plt.ylim([-12.5,5.5])
plt.grid(alpha=0.3)
plt.plot([-12,5],[-12,5], ls="--", c=".3", alpha=0.5)
plt.title("Cross-validation: Brain scores")
plt.text(-10, 3,'Pearson\'s r=0.83', ha='left', va='center')
plt.setp(ax.collections[1], alpha=0)
plt.savefig("Brain_LOOCV.png", dpi=300, bbox_inches = 'tight', pad_inches = 0.1)
```

```python
np.corrcoef(Lx_full, Lx_CV)[0,1]
```

```python
plt.figure()
ax=sns.regplot(x=Ly_full, y=Ly_CV, color="b", x_ci=0, fit_reg = 1, scatter_kws={"color": "black"}, line_kws={"color": "royalblue"});
plt.xlabel("Outcome scores (full)")
plt.ylabel("LOOCV Outcome scores")
plt.xlim([-2.62,1.5])
plt.ylim([-2.62,1.5])
plt.grid(alpha=0.3)
plt.plot([-2.5,2],[-2.5,2], ls="--", c=".3", alpha=0.5)
plt.title("Cross-validation: Outcome scores")
plt.text(-2, 0.95,'Pearson\'s r=0.99', ha='left', va='center')
plt.setp(ax.collections[1], alpha=0)
plt.savefig("Outcome_LOOCV.png", dpi=300, bbox_inches = 'tight', pad_inches = 0.1)
```
